# List available skills

```python
from dotenv import load_dotenv
import os
import anthropic

# Load .env from project root
load_dotenv("../.env")
print(f"API key loaded: {'Yes' if os.environ.get('ANTHROPIC_API_KEY') else 'No'}")

# Create client and list skills
client = anthropic.Anthropic()
skills = client.beta.skills.list(source="anthropic", betas=["skills-2025-10-02"])

print(f"Found {len(skills.data)} skills:")
for skill in skills.data:
    print(f"  {skill.id}: {skill.display_title}")
```
